```python
colors=['red','orange','blue']
varieties=['Setosa',
'Versicolor',
'Virginica'	]
```

```python
for i in range(3):
  x=df[df['variety']==varieties[i]]
  plt.scatter(x['sepal.length'],x['sepal.width'],c=colors[i],label=varieties[i])
  plt.xlabel('Sepal Length')
  plt.ylabel('Sepal Width')
  plt.legend()
```

```python
for i in range(3):
  x=df[df['variety']==varieties[i]]
  plt.scatter(x['petal.length'],x['petal.width'],c=colors[i],label=varieties[i])
  plt.xlabel('Petal Length')
  plt.ylabel('Petal Width')
  plt.legend()
  # we can conside this as bet because easily separable
```
